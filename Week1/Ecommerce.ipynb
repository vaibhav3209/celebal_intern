### Outcomes
| Observation                                                      | Actions                   |
|------------------------------------------------------------------|---------------------------|
| 1. Currency column has every value == 'INR'.                     | Removing Currency column. |
| 2. Final Price is a string column.                               | Convert to Integer.       |
| 3. Discount values are NULL.                                     | Fill with zero.           |
| 4. Encodings are different for (seller_name,seller_information). | Remove column.            |
| 5. Variations column contain empty braces.                       | Remove column.            |
| 6. Videos have the most null values.                             | Remove column.            |
| 7. Best offer columns has {} for null values.                    | Replace with NULL.        |


```python
# Drop these Columns
df_dropped= df.drop(['currency','seller_name','seller_information','variations','videos'],axis=1)


# fill discount Nan values with zero
df_dropped["discount"] = df_dropped["discount"].fillna(0).astype(float)

# Change Data type
df_dropped["final_price"] = (
    df_dropped["final_price"]
    .str.replace("₹", "", regex=False)
    .str.replace("\"", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

# Replace {} in best_offer column
df_dropped["best_offer"] = df_dropped["best_offer"].replace("{}", pd.NA)
```
